## Using ChemProp D-MPNN models

```python
from chemprop.train import run_hyperopt

args = [
    '--data_path', '../data/tox21.csv',
    '--dataset_type', 'classification',
    '--save_dir', 'tox21_opt_model',
    '--target_columns', 'NR-AR', 'NR-AR-LBD', 'NR-AhR', 'NR-Aromatase', 
                        'NR-ER', 'NR-ER-LBD', 'NR-PPAR-gamma',
                        'SR-ARE', 'SR-ATAD5', 'SR-HSE', 'SR-MMP', 'SR-p53',
    '--metric', 'auc',
    '--num_iters', '20',             # how many hyperparam configs to try
    '--config_save_path', 'best_hyperparams.json'
    '--split_type', 'scaffold', 
]

run_hyperopt(args)
```

```python
import pandas as pd

tox21 = pd.read_csv('../data/tox21.csv')
tox21_cleaned = tox21.drop(columns=['mol_id'])
tox21_cleaned
```

```python
tox21_cleaned.to_csv('../data/tox21_cleaned.csv', index=False)
```

```python
chemprop hpopt --data-path tox21_cleaned.csv \
    --task-type classification \
    --search-parameter-keywords depth ffn_num_layers message_hidden_dim learning_rate \
    --hpopt-save-dir results
```
