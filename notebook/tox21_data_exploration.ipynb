```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('../data/tox21.csv')
df
```

```python
# count number of missing row in each column and make a pandas table presenting this
missing = df.isnull().sum()
missing = pd.DataFrame(missing)
missing = missing[missing[0] > 0]
missing.columns = ['missing']
# missing.columns = ['column', 'missing']
missing['missing'] = missing['missing'].astype(int)
missing['percentage'] = missing['missing'] / len(df) * 100
missing = missing.sort_values(by='percentage', ascending=False)
missing = missing.reset_index()

missing 
```

```python
# compute the percentage of missing values
missing['missing'].sum() / (8006 * 12)
```

```python
# make a table to present how imbalance the dataset each for each of the 12 receptors
targets = ['NR-AR', 'NR-AR-LBD', 'NR-AhR', 'NR-Aromatase', 'NR-ER', 
           'NR-ER-LBD', 'NR-PPAR-gamma', 'SR-ARE', 'SR-ATAD5', 
           'SR-HSE', 'SR-MMP', 'SR-p53']
# make a table to present how imbalance the dataset each for each of the 12 receptors
imbalance = pd.DataFrame()
for target in targets:
    imbalance[target] = df[target].value_counts()
    imbalance[target] = imbalance[target].astype(int)
imbalance = imbalance.T
imbalance = imbalance.reset_index()
imbalance.columns = ['target', 'inactive', 'active']
imbalance['active percentage'] = imbalance['active'] / (imbalance['active'] + imbalance['inactive'])
imbalance = imbalance.sort_values(by='active percentage', ascending=False)
imbalance = imbalance.reset_index()
imbalance.drop(columns=['index'], inplace=True)

imbalance
```
